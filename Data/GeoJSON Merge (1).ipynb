```python
import geopandas as gp
```

```python
import pandas as pd
import numpy as np
```

```python
world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))
```

```python
world
```

```python
pollution = pd.read_csv('mismanaged-plastic-waste.csv')
```

```python
pollution = pollution.rename(columns = {'Code': 'iso_a3'})
pollution
```

```python
data = world.merge(pollution, on = 'iso_a3')
data
```

```python
data.to_file("countries_plastic_pollution.geojson", driver='GeoJSON')
```
