```python
import pandas as pd
import numpy as np
import os
```

```python
pd.read_csv('mmsi-daily-csvs-10-v2-2012/2012-01-01.csv')
```

```python
def process_file(csv):
    df = pd.read_csv(csv)
    return df
```

```python
files = os.listdir(os.getcwd())
folders = [file for file in files if file.startswith("mmsi-daily")]
print(folders)

dfs = []
for folder in folders:
    directory = os.fsencode(folder)
    for file in os.listdir(directory):
        filebytes = os.path.join(directory, file)
        fullfile = str(filebytes)[2:-1]
        dfs.append(process_file(fullfile))
```

```python
df = pd.concat(dfs)
```
